```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_excel('dataset/twitter_dialog.xlsx', header=None)
```

```python
data.head()
```

```python
reshaped_data = data.values.reshape(-1,)
```

```python
all_data = pd.Series(reshaped_data).dropna().to_numpy()
```

```python
all_data[0]
```

```python
kkma.morphs(all_data[0])
```

```python
def make_morphs(data):
    processed = list()
    
    for i, sen in enumerate(data):
        try: 
            processed.append(kkma.morphs(sen))
        except:
            print(f"number {i} / '{sen}' is passed")
            continue
            
    return processed
```

```python
processed = make_morphs(all_data)
```

```python
df = pd.DataFrame(processed)
```
